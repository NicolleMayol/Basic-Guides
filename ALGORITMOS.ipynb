### 1.2. Búsqueda binaria o Binary Search

La búsqueda binaria es un algoritmo más eficiente pero, sólo se puede aplicar a colecciones ordenadas. Funciona dividiendo repetidamente el espacio de búsqueda por la mitad. Compara el elemento intermedio con el valor objetivo y elimina la mitad del espacio de búsqueda según el resultado.

#### 1.2.1 Pseudo código:

izq = 0<br>
der =  tamaño_coleccion-1

<b>MIENTRAS</b> izq <= der <b> ENTONCES</b>:  <br>
&emsp;mitad = (izq+der)/2<br>
&emsp;<b>SI</b> colección[mitad] == elemento <b>ENTONCES:</b><br>
&emsp;&emsp;<b>RETORNAR</b> mitad<br>
&emsp;<b>SINO SI</b> colección[mitad] < elemento <b>ENTONCES:</b><br>
&emsp;&emsp; izq = mitad +1<br>
&emsp;<b>SINO:</b><br>
&emsp;&emsp; der = mitad -1<br>
<b>FIN MIENTRAS<br>
RETORNAR </b>-1<br>

#### 1.2.2 Código python:

```python
def binary_search(elemento, coleccion):
    izq = 0
    der = len(coleccion)-1

    while izq <= der:
        mitad = (izq + der) //2
        if elemento == coleccion[mitad]:
            return mitad
        elif elemento > coleccion[mitad]:
            izq = mitad + 1
        else:
            der = mitad -1
    return -1
```

#### 1.2.3 Eficiencia:

La búsqueda binaria reduce drásticamente el espacio de búsqueda con cada comparación, lo que la hace muy eficiente. Sin embargo, sólo funciona en colecciones ordenadas.

En términos de eficiencia de tiempo: O(log n)
